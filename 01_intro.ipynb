# Transformacoes

## Adding columns

```python
dfSpark.show()
```

```python
dfSpark.withColumn('experience_after_2_years',dfSpark['Experience']+2)
```

```python
dfSpark.show()
```

```python
dfSpark.withColumn('experience_after_2_years',dfSpark['Experience']+2).show()
```

```python
dfSpark.withColumn('age_by_experience',dfSpark['age']/dfSpark['Experience']).show()
```

```python
dfSpark = dfSpark.withColumn('age_by_experience',dfSpark['age']/dfSpark['Experience'])
```

```python
dfSpark.show()
```

```python
dfSpark
```

## drop columns

```python
dfSpark = dfSpark.drop('age_by_experience')
```

```python
dfSpark.show(5)
```

## Rename columns

```python
dfSpark = dfSpark.withColumnRenamed('Name','new_name')
```
